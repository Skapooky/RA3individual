# Questão 1 (vale até 2 pontos)
Determine a área entre as curvas $$y=xe^{-x^{2}}$$ $$y=x+1$$ $$x=2$$ e o eixo $y$, usando somente matplotlib, numpy, sympy. Trace o gráfico desta área preenchendo a área entre as curvas em uma cor diferente.



```python
import sympy as sp
import numpy as np
import matplotlib.pyplot as plt

x = sp.Symbol('x')
f = -x + x * sp.exp(-x**2) - 1

intersecao = sp.nsolve(f, x, [0, 2])

x = np.linspace(-3, 3, 100)
y1 = x*np.exp(-x**2)
y2 = x + 1
y3 = x - x + 2


plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.grid()
plt.fill_between(x, y1, y2, alpha=0.1, where=(x < intersecao), color="blue")
plt.fill_between(x, y2, y3, alpha=0.2, where=(x >= 1) & (y2 > 2), color="orange")
plt.fill_between(x, y1, y3, alpha=0.2,color="green")
```

# Questão 3 (vale até 4 pontos)
Calcule o valor das integrais definidas explicitadas a seguir. Trace o gráfico das funções que foram integradas e das funções encontradas após a integração destacando, com linhas verticais, os limites de integração.

##a)
$$\int_0^2 x^2+1 \space dx$$

##b)
$$\int_{-3}^1 6x^2-5x+2 \space dx$$



#A

```python
x = np.linspace(-1, 3, 100)
y = x**2 + 1

xIntegral = np.linspace(-1, 3, 100)
yIntegral = (x**3/3)+x

plt.plot(x, y, label='$f(x)= x^2 + 1$')
plt.title("Não integrada")
plt.fill_between(x, y, where=(x >= 0) & (x <= 2), alpha=0.2)
plt.legend()
plt.axvline(x = 0, color = "black")
plt.axvline(x = 2, color = "black")
plt.show()

plt.plot(xIntegral, yIntegral, label='$f(x)= \frac{x^3}{3} + x$')
plt.title("Integrada")
plt.fill_between(xIntegral, yIntegral, where=(x >= 0) & (x <= 2), alpha=0.2)
plt.legend()
plt.axvline(x = 0, color = "black")
plt.axvline(x = 2, color = "black")
plt.show()
```

#Questão 5 (vale até 1 ponto)
Encontre o valor da constante C para a antiderivada $F(x)$ da função $f(x) = 3x^3 - 2x + 1$ de tal forma que $F(1)=0$. Apresente seus cálculos em latex e explique as decisões que tomou.

Para achar a antiderivada precisamos fazer a integração da função f(x)
$$∫f(x) dx$$
$$∫3x^3 -2x +1$$
Usamos a regra da soma ou diferença
$$\int 3x^3 dx -  \int 2x dx+ \int 1 dx$$
Resolvemos as integrais usando a regra da potência
$$\int 3x^3 -  \int 2x + \int 1 $$
Isso nos dará as antiderivadas mais a constante C
$$\frac{3x^4}{4} - x^2 + x + C$$
Agora substituímos x por 1 e isso terá que resultar 0 de acordo com o enunciado
$$\frac{3\times1^4}{4} - 1^2 + 1 + C = 0$$
$$\frac{3}{4} - 1 + 1 + C = 0$$
$$\frac{3}{4} + C = 0$$
$$C = -\frac{3}{4}$$
